```python
import pandas as pd
from pathlib import Path
```

```python
entrada = Path("invoices_customer.csv")
df = pd.read_csv(entrada, sep=",")
```

```python
assert df['total_price'].notna().all(), "Hay ventas nulas."
assert pd.to_datetime(df['invoice_date'], errors='coerce').notna().all(), "Fechas inválidas."
# Tipificación​
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
df['mes'] = df['invoice_date'].dt.to_period('M').astype(str)
```

```python
# Exportar a Parquet (formato columnar apto para analítica)​
salida = Path("output/curated/ventas_clean.parquet")
df.to_parquet("salida.parquet", engine="pyarrow")
df.to_csv("salida_customer.csv", index=False)
```
